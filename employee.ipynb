```python
#Removing some column
df=df.drop('Over18',axis=1)
df=df.drop('EmployeeNumber',axis=1)
df=df.drop('StandardHours',axis=1)
df=df.drop('EmployeeCount',axis=1)
```

```python
#Transform the data
#Transform the categorical data into numerical data
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
    if df[column].dtype == 'np.number':
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

```python
#Create a new column
df['Age_Years'] = df['Age']
```

```python
#Drop the age column
df.drop('Age',axis=1,inplace=True)
```

```python
#Show the new column
df
```

```python
#Split the data into features and target variable
X=df.iloc[:,1:df.shape[1]].values
Y=df.iloc[:,0].values
```

```python
#Split the data into 75% training and 25% testing data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
#Use the decision tree classifier to train the model    
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(criterion='entropy',random_state=0)
tree.fit(X_train,Y_train)
```

```python
#Get the accuracy on the testing dataset
tree.score(X_test,Y_test)
```

```python
#Show the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,tree.predict(X_test))
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
print(cm)

print('Model_Testing_Accuracy={}'.format((TP+TN)/(TP+TN+FP+FN)))
```
